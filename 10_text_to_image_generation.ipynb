## Introduction to DALL-E-3: Text to Image Model

DALL-E-3 is an image generation behind GPT-4o.

Take note that each image cost $0.4, so don't go crazy in using image generation.


```python
# Imports
import base64
from io import BytesIO
from PIL import Image

import os
from dotenv import load_dotenv
from openai import OpenAI
```

```python
load_dotenv()
api_key = os.getenv("OPENAI_API_KEY")

openai = OpenAI(api_key=api_key)
openai
```

```python

def artist(city: str):
  image_reponse = openai.images.generate(
    model="dall-e-3",
    prompt=f"An image representing a vacation in {city} showing tourist sposts and everything unique about {city} in a vabriant pop-art style.",
    size="1024x1024",
    n=1,
    response_format="b64_json"
  )

  image_base64 = image_reponse.data[0].b64_json
  image_data = base64.b64decode(image_base64)
  return Image.open(BytesIO(image_data))
```

```python
image = artist("Philippines")
display(image)
```
